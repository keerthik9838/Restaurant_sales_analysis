# Objective : 
A restaurant in USA has limited presence accross a few states. Help the sales manager to analyze last 6 months data 

# Importing pandas library

```python
import pandas as pd 
```

```python
# Loading Data
data = pd.read_csv('Restaurant_sales_analysis.csv')
data
```

```python
# Problem 1 Sort the data in descending order of sales and display the result
```

```python
data.sort_values(by='Sales',ascending=False)
```

```python
# The list of sales in descending order of sales . 
```

```python
# Problem 2 Display the top 3 states with highest number of sales 
```

```python
Top_5_states = data.sort_values(by='Sales')
Top_5_states = data[['State','Sales']].head(3).sort_values(by='Sales',ascending=False)
Top_5_states
```

```python
#The Texas state has highest sales 
```

```python
# Problem 3 Display the states where sales  >= 45,000
```

```python
sales_45000 = data['Sales'] > 45000
data[sales_45000].sort_values(by='Sales',ascending=False)
```

```python
# California , Texas , Montana states has sales more than 45000$ 
```

```python
# Problem 4 Display the state wise total sales 
state_wise_sales = data.groupby(['State'])['Sales'].sum()
state_wise_sales.sort_values(ascending=False)
```

```python
# The Texas state has highest number of sales i.e 214856$
```

```python
# Problem 5 Display the month sales sales for the state with highest sales
top = data.groupby(['State','Month'])['Sales'].sum()
top['Texas'].sort_values(ascending = False)
```

```python
# In Texas state has highest number of sales in the month April i.e 45919$ and  lowest number of sales in mar 27215
```

```python
# Problem 6 Display the state and month with least sales 
```

```python

top_df = pd.DataFrame(top)
top_df['Sales'].sort_values()
```

```python
# The Alaska state in the month of  January has the least sales .
```

# Conclusion : 
# The Texas state has highest sales and in month of April there are more sales with respect to other months
